```python
import pandas as pd
import os

# Using double backslashes or raw string
file_path = r'C:\Users\Abdul\OneDrive - Chulalongkorn University\Data Basah\20240430_databasah_csv.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
#Buat data baru tanpa kolom yang tidak diperlukan aka numerical data
data_new = data.drop(['countrycode','provincecode', 'ftirrun_mmm','samplecode','Datapoints'], axis=1)
data_new.head(11)
```

```python
# Convert all values to numeric, coercing errors
data_new = data_new.apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values (if applicable)
data_new = data_new.dropna()

# Check the shape of the data after cleaning
data_shape = data_new.shape
print(f"Data shape after cleaning: {data_shape}")

def apply_snv(data):
    """Apply Standard Normal Variate (SNV) transformation to the dataset."""
    # Initialize an empty list to hold the transformed data
    snv_data = []

    # Iterate over each row (spectrum) in the dataset
    for spectrum in data:
        mean = np.mean(spectrum)
        std_dev = np.std(spectrum)
        snv_spectrum = (spectrum - mean) / std_dev
        snv_data.append(snv_spectrum)
    
    # Convert the list back to a NumPy array
    return np.array(snv_data)

# Convert the DataFrame to a NumPy array
data_array = data_new.values

# Apply the SNV transformation
snv_transformed_data = apply_snv(data_array)

# Convert back to DataFrame if needed
snv_transformed_df = pd.DataFrame(snv_transformed_data, columns=data_new.columns)

# Display the transformed data
snv_transformed_df.head()
```

```python

from sklearn.preprocessing import MinMaxScaler  # Importing MinMaxScaler
# Convert all values to numeric, coercing errors
data_new = data_new.apply(pd.to_numeric, errors='coerce')

# Drop rows with any NaN values (if applicable)
data_new = data_new.dropna()

# Check the shape of the data after cleaning
data_shape = data_new.shape
print(f"Data shape after cleaning: {data_shape}")

def apply_snv(data):
    """Apply Standard Normal Variate (SNV) transformation to the dataset."""
    # Initialize an empty list to hold the transformed data
    snv_data = []

    # Iterate over each row (spectrum) in the dataset
    for spectrum in data:
        mean = np.mean(spectrum)
        std_dev = np.std(spectrum)
        snv_spectrum = (spectrum - mean) / std_dev
        snv_data.append(snv_spectrum)
    
    # Convert the list back to a NumPy array
    return np.array(snv_data)

# Convert the DataFrame to a NumPy array
data_array = data_new.values

# Apply the SNV transformation
snv_transformed_data = apply_snv(data_array)

def apply_savgol(data, window_length=11, polyorder=2):
    """Apply Savitzky-Golay filter to the dataset."""
    return savgol_filter(data, window_length=window_length, polyorder=polyorder, axis=1)

# Apply the Savitzky-Golay filter
savgol_transformed_data = apply_savgol(snv_transformed_data)

def apply_range_scaling(data):
    """Apply Range Scaling (Min-Max Scaling) to the dataset."""
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data)
    return scaled_data

# Apply the Range Scaling
range_scaled_data = apply_range_scaling(savgol_transformed_data)

# Convert back to DataFrame if needed
range_scaled_df = pd.DataFrame(range_scaled_data, columns=data_new.columns)

# Display the transformed data
range_scaled_df.head()
```
