```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

# Assuming `data_new` is your DataFrame with 11 rows and 3369 columns
# For example purpose, generating random data
data_new = pd.DataFrame(np.random.rand(11, 3369))

# Step 1: Convert DataFrame to 2D Array
data_array = data_new.to_numpy()

# Step 2: Apply Polynomial Features
degree = 2  # You can choose the degree of polynomial features
poly = PolynomialFeatures(degree)
poly_features = poly.fit_transform(data_array)

# Convert the resulting array back to a DataFrame to view as a table
poly_features_df = pd.DataFrame(poly_features)

# Display the DataFrame
poly_features_df.head(11)
```
